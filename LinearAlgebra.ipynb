Adding and subtracting two vectors without numpy

```python
def vector_add(v, w):
    """adds corresponding elements"""
    return [v_i + w_i for v_i, w_i in zip(v, w)]
```

```python
def vector_subtract(v, w):
    """subtracts corresponding elements"""
    return [v_i - w_i for v_i, w_i in zip(v, w)]
```

 componentwise sum a list of vectors. That is, create a
new vector whose first element is the sum of all the first elements, whose second eleâ€
ment is the sum of all the second elements, and so on. 

```python
#The easiest way to do this is by adding one vector at a time:

def vector_sum(vectors):
    """sums all corresponding elements"""
    result = vectors[0]  # start with the first vector
    # print(result)
    for vector in vectors[1:]: # then loop through the others
        result = vector_add(result, vector) # and add them one at a time
        # print(result)
    return result

vectors = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

vector_sum(vectors)  # => [12, 15, 18]
```

```python
def scalar_multiply(c, v):
    """multiplies every element by c"""
    return [c * v_i for v_i in v]

def vector_mean(vectors):
    """computes the element-wise average"""
    n = len(vectors)
    return scalar_multiply(1/n, vector_sum(vectors)) 
```

```python
vectors = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
vector_mean(vectors)  # => [4.0, 5.0, 6.0]
```

Dot product

```python
def dot(v, w):
    """computes the dot product of v and w"""
    return sum(v_i * w_i for v_i, w_i in zip(v, w)) # scalar product and sum
```

```python
def sum_of_squares(v):
    """computes the sum of squares of v"""
    return dot(v, v)
```

```python
import math
def magnitude(v):
    """computes the magnitude (length) of v"""
    return math.sqrt(sum_of_squares(v))

# we can use squared distance too by passing Zero vector as one argument
```

```python
def squared_distance(v, w):
    """computes the squared distance between v and w"""
    return sum_of_squares(vector_subtract(v, w))

def distance(v, w):
    """computes the distance between v and w"""
    return math.sqrt(squared_distance(v, w))

```

##### Matrices

```python
def shape(A):
    num_rows = len(A)
    num_cols = len(A[0]) if A else 0 # number of elements in first row
    return num_rows, num_cols
```

```python
def get_row(A, i):
    return A[i] # A[i] is already the ith row

def get_column(A, j):
    return [A_i[j] # jth element of row A_i
    for A_i in A] # for each row A_i
```
