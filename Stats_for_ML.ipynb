Plot Normal/Bernoulli distributions with Matplotlib


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(-4, 4, 400)
x

mu = 0
sigma = 1
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2) # PDF of normal distribution
```

```python
plt.plot(x, pdf)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title("Normal Distribution")
plt.show()
```

```python
# Bernoulli Distribution

p = 0.6
x = [0, 1]
y = [1 - p, p]
```

```python
plt.bar(x, y)
plt.xticks([0, 1])
plt.xlabel("x")
plt.ylabel("Probability")
plt.title("Bernoulli Distribution")
plt.show()
```
